```python
max_goals = df['goals'].max()
min_goals = df['goals'].min()
mean_assists = df['assists'].mean()
median_minutes = df['minutes_played'].median()

print("Maximum Goals:", max_goals)
print("Minimum Goals:", min_goals)
print("Average Assists:", mean_assists)
print("Median Minutes Played:", median_minutes)
```

```python
sorted_players = df.sort_values(by='goals', ascending=False)

sorted_players[['player_name', 'goals']].head(10)
```

```python
top_scorers = df[df['goals'] > 2]

top_scorers[['player_name', 'goals']]
```

```python
import matplotlib.pyplot as plt

top10 = df.sort_values(by='goals', ascending=False).head(10)

plt.figure(figsize=(10,5))
plt.bar(top10['player_name'], top10['goals'])

plt.title("Top 10 Goal Scorers")
plt.xlabel("Player Name")
plt.ylabel("Goals")

plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['goals'])

plt.title("Distribution of Goals")
plt.xlabel("Goals")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['goals'], df['assists'])

plt.xlabel("Goals")
plt.ylabel("Assists")
plt.title("Goals vs Assists")

plt.show()
```
